```python
import re

import pandas as pd
import numpy as np
pd.options.display.max_rows = None

def make_clickable(val):
    return '<a href="{}">{}</a>'.format(val, val)
```

```python
df = pd.read_csv('group.csv')
df = df.sort_values(by=['price_max'])
df = df.drop_duplicates(subset=['itemid'])
df = df.reset_index(drop=True)
print('Total    :', len(df))
print('Max price:', df['price_max'].max())
df.head()
```

```python
show_cols = ['name', 'price_max', 'link']
codes = [
    'h61', 'p67', 'q67', 'z68',
    'h71', 'b75', 'h77', 'p77', 'q77', 'z77',
    'h81', 'b85', 'h87', 'z87',
    'h97', 'z97']
for code in codes:
    key_pattern = re.compile("%s-?(i|n|(m\-itx))" % code, re.IGNORECASE)
    ignore = re.compile("故障", re.IGNORECASE)
    included = df['name'].apply(lambda name: key_pattern.search(name) != None and ignore.search(name) == None)
    result = df[included == True][show_cols]
    if len(result) != 0:
        result = result.style.format({'link': make_clickable})
        print(code)
        display(result)
```
